#drop

```python
for i in range(len(file_list)):
    print(file_list[i])
    df = pd.read_csv('/content/drive/MyDrive/rehashtag525/' + file_list[i] + '/' + file_list[i] + '.csv', encoding='utf-8-sig', index_col=0)
    df = df.astype(str)
    for j in range(len(df)):
      if df['images'][j] == '[]':
        df.drop(j, inplace=True)
      elif df['hashtags'][j] == '[]':
        df.drop(j, inplace=True)
      elif df['location_info'][j] == 'nan':
        df.drop(j, inplace=True)

    df = df.reset_index()
    df.to_csv('/content/drive/MyDrive/rehashtag525/' + file_list[i] + '/' + file_list[i] + '_drop_0530.csv', encoding='utf-8-sig')
```
