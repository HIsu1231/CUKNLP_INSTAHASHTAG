```python
for i, texts in enumerate(df['images']):
    try:
        images = texts.split(',') 
        for j in range(len(images)):
            image = images[j][2:-2]
            with urlopen(image) as f:
                with open('./img/' + str(i) + '-' + str(j) + '.jpg','wb') as h:
                    img = f.read()
                    h.write(img)
    except:
        continue
    # print(i,images)
```
